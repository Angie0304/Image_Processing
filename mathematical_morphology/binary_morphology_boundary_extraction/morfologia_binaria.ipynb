```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen 1 en escala de grises
imagen = cv2.imread('data/hombre_binaria.bmp', 0)
# Mostrar la imagen
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original 1')
plt.show()
```

```python
# Cargar la imagen 2 en escala de grises
imagen2 = cv2.imread('data/rostro_mujer.bmp', 0)
# Mostrar la imagen
plt.imshow(imagen2, cmap='gray')
plt.title('Imagen Original 2')
plt.show()
```

```python
# Cargar la imagen 3 en escala de grises
imagen3 = cv2.imread('data/perro.bmp', 0)
# Mostrar la imagen
plt.imshow(imagen3, cmap='gray')
plt.title('Imagen Original 3')
plt.show()
```

```python
# Cargar la imagen 4 en escala de grises
imagen4 = cv2.imread('data/arbol.bmp', 0)
# Mostrar la imagen
plt.imshow(imagen4, cmap='gray')
plt.title('Imagen Original 4')
plt.show()
```

Paso 2. Erosión

```python
# Definir el kernel para la erosión
kernel = np.ones((5,5), np.uint8)
# Aplicar la operación de erosión
imagen_erosionada = cv2.erode(imagen, kernel, iterations = 1)
# Mostrar la imagen erosionada
plt.imshow(imagen_erosionada, cmap='gray')
plt.title('Imagen Erosionada 1')
plt.show()
```

```python
imagen_erosionada2 = cv2.erode(imagen2, kernel, iterations = 1)
# Mostrar la imagen erosionada
plt.imshow(imagen_erosionada2, cmap='gray')
plt.title('Imagen Erosionada 2')
plt.show()
```

```python
imagen_erosionada3 = cv2.erode(imagen3, kernel, iterations = 1)
# Mostrar la imagen erosionada
plt.imshow(imagen_erosionada3, cmap='gray')
plt.title('Imagen Erosionada 3')
plt.show()
```

```python
imagen_erosionada4 = cv2.erode(imagen4, kernel, iterations = 1)
# Mostrar la imagen erosionada
plt.imshow(imagen_erosionada4, cmap='gray')
plt.title('Imagen Erosionada 4')
plt.show()
```

Paso 3. Dilatación

```python
# Aplicar la operación de dilatación
imagen_dilatada = cv2.dilate(imagen, kernel, iterations = 1)
# Mostrar la imagen dilatada
plt.imshow(imagen_dilatada, cmap='gray')
plt.title('Imagen Dilatada 1')
plt.show()
```

```python
imagen_dilatada2 = cv2.dilate(imagen2, kernel, iterations = 1)
# Mostrar la imagen dilatada
plt.imshow(imagen_dilatada2, cmap='gray')
plt.title('Imagen Dilatada 2')
plt.show()
```

```python
imagen_dilatada3 = cv2.dilate(imagen3, kernel, iterations = 1)
# Mostrar la imagen dilatada
plt.imshow(imagen_dilatada3, cmap='gray')
plt.title('Imagen Dilatada 3')
plt.show()
```

```python
imagen_dilatada4 = cv2.dilate(imagen4, kernel, iterations = 1)
# Mostrar la imagen dilatada
plt.imshow(imagen_dilatada4, cmap='gray')
plt.title('Imagen Dilatada 4')
plt.show()
```

Paso 5. Apertura

```python
# Aplicar la operación de apertura
imagen_apertura = cv2.morphologyEx(imagen, cv2.MORPH_OPEN, kernel)
# Mostrar la imagen con apertura
plt.imshow(imagen_apertura, cmap='gray')
plt.title('Imagen con Apertura 1')
plt.show()
```

```python
imagen_apertura2 = cv2.morphologyEx(imagen2, cv2.MORPH_OPEN, kernel)
# Mostrar la imagen con apertura
plt.imshow(imagen_apertura2, cmap='gray')
plt.title('Imagen con Apertura 2')
plt.show()
```

```python
imagen_apertura3 = cv2.morphologyEx(imagen3, cv2.MORPH_OPEN, kernel)
# Mostrar la imagen con apertura
plt.imshow(imagen_apertura3, cmap='gray')
plt.title('Imagen con Apertura 3')
plt.show()
```

```python
imagen_apertura4 = cv2.morphologyEx(imagen4, cv2.MORPH_OPEN, kernel)
# Mostrar la imagen con apertura
plt.imshow(imagen_apertura4, cmap='gray')
plt.title('Imagen con Apertura 4')
plt.show()
```

Paso 6. Cierre

```python
# Aplicar la operación de cierre
imagen_cierre = cv2.morphologyEx(imagen, cv2.MORPH_CLOSE, kernel)
# Mostrar la imagen con cierre
plt.imshow(imagen_cierre, cmap='gray')
plt.title('Imagen con Cierre')
plt.show()
```

```python
imagen_cierre2 = cv2.morphologyEx(imagen2, cv2.MORPH_CLOSE, kernel)
# Mostrar la imagen con cierre
plt.imshow(imagen_cierre2, cmap='gray')
plt.title('Imagen con Cierre 2')
plt.show()
```

```python
imagen_cierre3 = cv2.morphologyEx(imagen3, cv2.MORPH_CLOSE, kernel)
# Mostrar la imagen con cierre
plt.imshow(imagen_cierre3, cmap='gray')
plt.title('Imagen con Cierre 3')
plt.show()
```

```python
imagen_cierre4 = cv2.morphologyEx(imagen4, cv2.MORPH_CLOSE, kernel)
# Mostrar la imagen con cierre
plt.imshow(imagen_cierre4, cmap='gray')
plt.title('Imagen con Cierre 4')
plt.show()
```

```python
gradiente = cv2.subtract(imagen_dilatada, imagen_erosionada)

# Mostrar la imagen resultante
plt.imshow(gradiente, cmap='gray')
plt.title('Frontera')
plt.show()
```

```python
imagen_resta2 = cv2.subtract(imagen2, imagen_erosionada2)

# Mostrar la imagen resultante
plt.imshow(imagen_resta2, cmap='gray')
plt.title('Frontera')
plt.show()
```

```python
imagen_resta3 = cv2.subtract(imagen3, imagen_erosionada3)

# Mostrar la imagen resultante
plt.imshow(imagen_resta3, cmap='gray')
plt.title('Frontera')
plt.show()
```

```python
imagen_resta4 = cv2.subtract(imagen4, imagen_erosionada4)

# Mostrar la imagen resultante
plt.imshow(imagen_resta4, cmap='gray')
plt.title('Frontera')
plt.show()
```
